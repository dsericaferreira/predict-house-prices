# Create a simple regression module of sqft_living to price!!

```python
training_data,test_data = sales.random_split(.8, seed=0)
```

```python
training_data
```

# Build the regression model:

```python
#Build a square foot prediction model - Linear regression!!

#sqr_ft = turicreate.linear_regression.create( dataset, target="", features =[''])
sqr_ft = turicreate.linear_regression.create(training_data, target="price", features=['sqft_living'])
```

# Evaluate the model:


```python
print(sqr_ft.evaluate(test_data))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(test_data['sqft_living'], test_data['price'], '.',
        test_data['sqft_living'], sqr_ft.predict(test_data), ',')
```

# Inspecting the model:
coefficients

### Building a new model:

#### Steps:

```python
houses_data = turicreate.SFrame('~/home_data.sframe')
houses_data.show()
plt.plot(houses_data['sqft_living'], houses_data['price'], '.')
```

```python
training, test = houses_data.random_split(.8, seed=0)
```

```python
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house				
'grade', # measure of quality of construction				
'waterfront', # waterfront property				
'view', # type of view				
'sqft_above', # square feet above ground				
'sqft_basement', # square feet in basement				
'yr_built', # the year built				
'yr_renovated', # the year renovated				
'lat', 'long', # the lat-long of the parcel				
'sqft_living15', # average sq.ft. of 15 nearest neighbors 				
'sqft_lot15', # average lot size of 15 nearest neighbors 
]

reg_model = turicreate.linear_regression.create(training, target = 'price', features = advanced_features, validation_set=None)
```

```python
print(reg_model.evaluate(test))
```

```python
plt.plot(test['sqft_living'], test['price'], '.',
        test['sqft_living'], reg_model.predict(test_data), ',')
```

```python
reg_model.coefficients
```

```python
my_features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
my_features_model = turicreate.linear_regression.create(training, target='price', features = my_features, validation_set=None)
```

```python
print(my_features_model.evaluate(test))
```

```python
print(my_features_model.evaluate(test)['rmse'] - reg_model.evaluate(test)['rmse'])
```
